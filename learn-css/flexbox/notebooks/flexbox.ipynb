
### Initial Setup


The [`basic.css`](./src/basic.css) file contains vanilla css styling unrelated to flexbox itself:


````python
/* ./src/basic.css */
.container > div {
  padding: 10px;
  text-align: center;
  font-size: 2em;
  color: #ffeead;
}

html,
body {
  background-color: #ffeead;
  margin: 10px;
}

.container > div:nth-child(1) {
  background-color: #96ceb4;
}

.container > div:nth-child(2) {
  background-color: #ff6f69;
}

.container > div:nth-child(3) {
  background-color: #88d8b0;
}

.container {
  border: 5px solid #ffcc5c;
}
````

### HTML Structure ([`index.html`](./src/00.html))


````python
<!-- index.html (./src/00-initial.html) -->
<!DOCTYPE html>
<html>
  <head>
    <link rel="stylesheet" href="basic.css" />
    <script src="https://cdn.jsdelivr.net/npm/@tailwindcss/browser@4"></script>
  </head>

  <body>
    <nav class="container">
      <div>Home</div>
      <div>Search</div>
      <div>Logout</div>
    </nav>
  </body>
</html>
````

The markup contains a `nav` element with the class `container`, and inside are three `div` elements which will become **flex items**.

![Initial layout](./assets/00-initial.png)

### Key Concepts

**Container and Items:** The fundamental structure of Flexbox requires:

- A **container** (parent element)
- **Items** (direct children of the container)

These don't have to be `div`s or `nav`s—they could be `ul` and `li` elements, or any other combination. As long as they are direct children of the Flexbox container, they become flex items.

### Activating Flexbox

Without any flexbox properties, the `div`s stack vertically by default.

To activate Flexbox, add `flex` to the container:

````python
<!-- index.html (./src/01-flex.html) -->
<!DOCTYPE html>
<html>
  <head>
    <link rel="stylesheet" href="basic.css" />
    <script src="https://cdn.jsdelivr.net/npm/@tailwindcss/browser@4"></script>
  </head>

  <body>
    <nav class="container flex">
      <div>Home</div>
      <div>Search</div>
      <div>Logout</div>
    </nav>
  </body>
</html>
````

![Flexbox activated](./assets/01-flex.png)

**What Changed:**

- Elements automatically align horizontally instead of vertically
- By default, Flexbox distributes items from left to right
- The Flexbox container is a block element, taking up all available width (minus margins)

## 2. Main Axis and Cross Axis

Understanding **axes** is crucial for working with Flexbox, as a Flexbox container always has a **direction**.

![Axis diagram](./assets/02-axis.png)

### Default Direction (Row)

By default, the flex-direction is horizontal:

- **Main axis:** left to right (along the row)
- **Cross axis:** top to bottom

````python
<!-- index.html (./src/03-flex-row.html) -->
<!DOCTYPE html>
<html>
  <head>
    <link rel="stylesheet" href="basic.css" />
    <script src="https://cdn.jsdelivr.net/npm/@tailwindcss/browser@4"></script>
  </head>

  <body>
    <nav class="container flex flex-row">
      <div>Home</div>
      <div>Search</div>
      <div>Logout</div>
    </nav>
  </body>
</html>
````

![Row direction](./assets/03-flex-row.png)

### Column Direction

You can flip the direction by changing `flex-direction` to `column`:

````python
<!-- index.html (./src/04-flex-col.html) -->
<!DOCTYPE html>
<html>
  <head>
    <link rel="stylesheet" href="basic.css" />
    <script src="https://cdn.jsdelivr.net/npm/@tailwindcss/browser@4"></script>
  </head>

  <body>
    <nav class="container flex flex-col">
      <div>Home</div>
      <div>Search</div>
      <div>Logout</div>
    </nav>
  </body>
</html>
````

![Column direction](./assets/04-flex-col.png)

Now:

- **Main axis:** top to bottom
- **Cross axis:** left to right

### Why This Matters

Different CSS properties control positioning along the main axis versus the cross axis. Understanding which axis is which determines which properties you'll use.

**Note:** This guide focuses primarily on `row` direction (the default), as it's more commonly used in practical website development.

![Default layout](./assets/01-flex.png)

## 3. Justify Content

The `justify-content` property positions items along the **main axis** (left to right by default, due to the `dir="ltr"` attribute in the HTML tag).

### Available Values

`justify-content` accepts these values:

- `flex-start` (default)
- `flex-end`
- `center`
- `space-around`
- `space-between`
- `space-evenly`

### Flex-Start (Default)

````python
<!-- index.html (./src/05-justify-start.html) -->
<!DOCTYPE html>
<html>
  <head>
    <link rel="stylesheet" href="basic.css" />
    <script src="https://cdn.jsdelivr.net/npm/@tailwindcss/browser@4"></script>
  </head>

  <body>
    <nav class="container flex flex-row justify-start">
      <div>Home</div>
      <div>Search</div>
      <div>Logout</div>
    </nav>
  </body>
</html>
````

![Flex-start](./assets/05-justify-start.png)

Content is squeezed together at the start of the main axis (left side). This is what you saw in the default flex layout.

### Flex-End

````python
<!-- index.html (./src/06-justify-end.html) -->
<!DOCTYPE html>
<html>
  <head>
    <link rel="stylesheet" href="basic.css" />
    <script src="https://cdn.jsdelivr.net/npm/@tailwindcss/browser@4"></script>
  </head>

  <body>
    <nav class="container flex justify-end">
      <div>Home</div>
      <div>Search</div>
      <div>Logout</div>
    </nav>
  </body>
</html>
````

![Flex-end](./assets/06-justify-end.png)

Content moves to the end of the main axis (right side).

### Space-Around

Each item gets equal space on its left and right sides. Notice that the space between two adjacent items is twice as wide as the space at the edges, because the spacing from each item combines.
